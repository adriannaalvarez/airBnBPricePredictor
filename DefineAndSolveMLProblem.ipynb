```python
#Changing the boolean columns with 0 and 1 so the model can understand
df = df.astype({col: 'int' for col in df.select_dtypes(include=['bool']).columns})
```

```python
#Checking if any of the object types can be one hot encoded
objects= df.select_dtypes(include=['object'])
objects.nunique()
```

```python
#One hot encodes the neighborhood group cleansed column so it will show a 1 or 0 for the neighborhood instead of Strings
df_neighborhood_group_cleansed= pd.get_dummies(df['neighbourhood_group_cleansed'], prefix='neighborhood_group_cleansed_')
df_neighborhood_group_cleansed
```

```python
##One hot encodes the room type column so it will show a 1 or 0 for the room type instead of Strings
df_room_type= pd.get_dummies(df['room_type'], prefix='room_type_')
df_room_type
```

```python
#Joins the one hot encoded columns and removes the columns they replaced
df= df.join(df_neighborhood_group_cleansed)
df.drop(columns='neighbourhood_group_cleansed',inplace=True)
df= df.join(df_room_type)
df.drop(columns='room_type',inplace=True)
```

```python
#Now that we changed the columns above to numerical values, we can remove 
df= df.select_dtypes(exclude=['object'])
```

```python
df.dtypes
```

```python
nan_count = np.sum(df.isnull(), axis = 0)
nan_detected = nan_count != 0
nan_detected
```

```python
#Replacing the missing values with the mean only makes sense for the columns that contain numerical values
is_int_or_float = (df.dtypes == np.int64) | (df.dtypes == np.float64)
to_impute = nan_detected & is_int_or_float
to_impute
```

```python
#these columns contain missing values that can be replaced with mean values
to_impute_selected = list(df.columns[to_impute])
to_impute_selected
```

```python
#For every column listed in `to_impute_selected`, fill the missing values with the corresponding mean of all values in the column
for column in to_impute_selected:
    df[column].fillna(value=df[column].mean(), inplace=True)

#Checking results
for colname in to_impute_selected:
    print("{} missing values count :{}".format(colname, np.sum(df[colname].isnull(), axis = 0)))
```

```python
df.nunique()
```

```python
#These features only have one value, so they need to be removed as they do not add any nuance to the data
to_remove= df.nunique() == 1
to_remove
```

```python
#This selects and deletes those features
to_remove_selected = list(df.columns[to_remove])
df= df.drop(columns=to_remove_selected)
```

```python
#Create a new column, titled `pricing` by winsorizing the ‘price’ column with the top and bottom 1% percentile values. Also price column is dropped
#to protect from data leakage. Pricing becomes new label 
import scipy.stats as stats
df['pricing'] = stats.mstats.winsorize(df['price'], limits=[0.1, 0.1])
df= df.drop(columns='price')
```

```python
print('Begin Random Forest Implementation...')
# 1. Create the RandomForestRegressor model with the best hyperparameters according to the param grid
best_rf = RandomForestRegressor(criterion='squared_error', n_estimators=100,max_depth=32,min_samples_leaf=5)

# 2. Fit the model to the training data below
best_rf.fit(X_train, y_train)

# 3. Make predictions on the X_test
best_rf_preds = best_rf.predict(X_test)
best_rf_preds

print('End')
```

```python
# The mean squared error for best_rf
print('\nModel Performance of best_rf\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, best_rf_preds)))
# The coefficient of determination: 1 is perfect prediction
print(' R^2 =   %.2f'
      % r2_score(y_test, best_rf_preds))
```

```python
#We tried to see what features held the most importance to the model using MDI method 
importances = best_rf.feature_importances_
std = np.std([tree.feature_importances_ for tree in best_rf.estimators_], axis=0)

feature_df = pd.DataFrame({
    'importance': importances,
    'std': std
}, index=list(features))
sorted_feature_df = feature_df.sort_values(by='importance', ascending=False)

fig, ax = plt.subplots()
sorted_feature_df['importance'].plot.bar(yerr=sorted_feature_df['std'], ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()

plt.show()
```

```python
#We tried to see what features held the most importance to the model using permutation method 
result = permutation_importance(
    best_rf, X_test, y_test, n_repeats=10, random_state=42, n_jobs=2
)

forest_importances = pd.Series(result.importances_mean, index=list(features))
sorted_importances = forest_importances.sort_values(ascending=False)
feature_indices = {feature: i for i, feature in enumerate(features)}
sorted_std = result.importances_std[[feature_indices[feature] for feature in sorted_importances.index]]

fig, ax = plt.subplots()
sorted_importances.plot.bar(yerr=sorted_std, ax=ax)
ax.set_title("Feature importances using permutation on full model")
ax.set_ylabel("Mean accuracy decrease")
fig.tight_layout()

plt.show()
```

```python
#K fold method to ensure that the rmse & R^2 scores do not change drastically, showing that it can 
# perform well with new data
num_folds = 5
folds = KFold(n_splits = num_folds, random_state=None)

rmse_scores = []
r2_scores = []

for train_row_index , test_row_index in folds.split(X_train): 
    
    # our new partition of X_train and X_val
    X_train_new  = X_train.iloc[train_row_index] 
    X_val = X_train.iloc[test_row_index]
    
    # our new partition of y_train and y_val
    y_train_new = y_train.iloc[train_row_index]
    y_val = y_train.iloc[test_row_index]
    
    best_rf.fit(X_train_new, y_train_new)
    predictions = best_rf.predict(X_val)
    cv = KFold(n_splits=5, shuffle=True, random_state=42)

    rmse = np.sqrt(np.mean((predictions - y_val) ** 2))
    rmse_scores.append(rmse)
    r2 = r2_score(y_val, predictions)
    r2_scores.append(r2)
     
        
for i in range(num_folds):
    print(f'Iteration {i+1} - RMSE: {rmse_scores[i]}, R²: {r2_scores[i]}')
    
print("Mean RMSE:", np.mean(rmse_scores))
print("Mean R²:", np.mean(r2_scores))
```
